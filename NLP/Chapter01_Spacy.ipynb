## Sentence Boundary Detection (SBD)

In NLP, sentence boundary detection, or SBD, is the identification of sentences in a text. Again, this may seem fairly easy to do with rules. One could use split(“.”), but in English we use the period to also denote abbreviation like Dr. MR. etc. You could, again, write rules to look for periods not proceeded by a lowercase word, but again, I ask the question, “why bother?”

```python
for sent in doc.sents:
    print(sent)

## see the U.S. is not splitted into two seperate things
```

```python
sentence1 = list(doc.sents)[0]
print(sentence1)

## U.S. is not splitted into two different entities
```
