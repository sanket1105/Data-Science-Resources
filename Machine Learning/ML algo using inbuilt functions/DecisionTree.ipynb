```python

import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd 
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../Datasets/Salaries.csv")
df.head()
```

```python
print(df['job'].value_counts())
print("")

print(df['degree'].value_counts())
print("")

print(df['company'].value_counts())
```

```python
## lets use label encoder on all
## infact we should use one hot encoder on degre as tehy are ordinal
## and label encoder on rest of tha all

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['company_le'] = le.fit_transform(df['company'])
df['job_le'] = le.fit_transform(df['job'])
df['degree_le'] = le.fit_transform(df['degree'])

df.head()
```

```python
df_new = df.drop(['company','job','degree'],axis=1)
df_new
```

```python
X = df_new.drop("salary_more_then_100k",axis=1)
y = df_new['salary_more_then_100k']
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()

model.fit(X,y)
model.score(X,y)
```
